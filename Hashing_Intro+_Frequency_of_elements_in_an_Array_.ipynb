# Naive Approach

image.png

````python
def calculate_freq(array,query):
    count = 0
    for j in range(0, len(array)):
        if array[j] == query:
            count += 1
    print(count)


array = [7,8,5,8,8,7,5,1,8]
calculate_freq(array, 8)
calculate_freq(array, 3)
````


```
### TC: O(N*Q)
### SC: O(1)


```





````python
n = int(input())
array = [int(input()) for _ in range(n)]
q = int(input())

for i in range(0, q):
    query = int(input())

    count = 0
    for j in range(0, n):
        if array[j] == query:
            count += 1
    print(count)
````

# Frequency Array

image.png

````python
n = int(input())
array = []
hash = [0] * 10

for i in range(0, n):
    array.append(int(input()))
    hash[array[i]] = hash[array[i]] + 1

q = int(input())

for i in range(0, q):
    query = int(input())
    count = hash[query]
    print(count)
````


```
Time Complexity: O(n+q), where n is the size of the given array
Auxiliary Space: O(n)


```

# Using a HashMap

image.png


```

Time Complexity: O(n), where n is the size of the given array



```

````python
from collections import defaultdict

n = int(input())
array = []
hash_map = defaultdict(int)

for i in range(0, n):
    num = int(input())
    array.append(num)
    hash_map[num] += 1

q = int(input())

for i in range(0, q):
    query = int(input())
    count = hash_map[query]
    print(count)

````
